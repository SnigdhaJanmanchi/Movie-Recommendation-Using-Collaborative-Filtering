```python
#Importing Necessary Libraries
import pandas as pd
from sklearn.metrics.pairwise import cosine_similarity
```

```python
#Reading datasets
df1 = pd.read_csv("/content/drive/MyDrive/Saved from Chrome/movies.csv")
df2 = pd.read_csv("/content/drive/MyDrive/Saved from Chrome/ratings.csv")
```

```python
df = df2.merge(df1, left_on='movieId', right_on='movieId', how='left')
df
```

```python
#Removing columns because we won't be using them
del df['timestamp']
del df['genres']
```

```python
df.head()
```

We are now going to transform the DF into a movie-to-user matrix, where each row represents one movie and the columns correspond to different users

```python
user_movie_matrix = pd.pivot_table(df, values = 'rating', index='movieId', columns = 'userId')
user_movie_matrix
```

```python
user_movie_matrix = user_movie_matrix.fillna(0)
user_movie_matrix.head()
```

Calculating similarites

Similarity based methods determine the most similar objects with the highest values as it implies they live in closer neighborhoods. There are multiple similarity based metrics

Pearsonâ€™s correlation

Spearmanâ€™s correlation

Kendallâ€™s Tau

Cosine similarity

Jaccard similarity

However, for our case we'll be using Pearsonâ€™s correlation because we have a sparse matrix, which means that most movies are not rated (have a rating of 0). hence we will center all the ratings to 0 so that the default rating becomes 0.


```python
#user-based collaborative filtering

user_user_matrix = user_movie_matrix.corr(method='pearson')
user_user_matrix
```

As we can see from the above matrix, higher +ve values means more similar and -ve values mean not similar. Eg: Similarty of user 1 and user 2 is 0.019400 but similarity of user1 and user3 is 0.053056, which means User1 is more similar to User3.

Recommending Movies for UserId 2

Movie list for userid 2

```python
#Extracting top 10 similar users for User2 by sorting them in descending order based on their similarties

user_user_matrix.loc[2].sort_values(ascending=False).head(10)
```

```python
#Converting the above data into a DF and removing the user2 itself

df_2 = pd.DataFrame(user_user_matrix.loc[2].sort_values(ascending=False).head(10))
df_2 = df_2.reset_index()
df_2.columns = ['userId', 'similarity']
```

```python
df_2 = df_2.drop((df_2[df_2['userId'] ==2]).index)
df_2
```

```python
#New DF which has the similar users and rated movies

final_df = df_2.merge(df, left_on='userId', right_on='userId', how='left')
final_df
```

We obtain a new metric (score) by combining how similar two people are and their movie rating. The basic thought is that a user who is most similar to you may suggest movies you would like best. If the most similar user gives a movie a high ranking, our algorithm will place it higher in the recommendation.

```python
final_df['score'] = final_df['similarity']*final_df['rating']
final_df
```

```python
#Creating a df for all the movies which are already watched by our target User 2

watched_df = df[df['userId'] == 2]
watched_df
```

In this Step we are removing the already watched movies because we don't want to recommend them again

```python
cond = final_df['movieId'].isin(watched_df['movieId'])
final_df.drop(final_df[cond].index, inplace = True)
```

```python
recommended_df = final_df.sort_values(by = 'score', ascending = False)['title'].head(10)
recommended_df = recommended_df.reset_index()
del recommended_df['index']
```
